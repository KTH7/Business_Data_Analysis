## 문제 발견

* 문제
  - '낚시랜드'는 다른 게임과 비교했을 때, 피처폰 유저수는 비슷하게 줄어든 반면, 스마트폰 유저수의 증가는 적었다.
* 피처폰 유저의 탈퇴를 구조적으로 분석
  1. 피처폰에서 그냥 탈퇴
  2. 피처폰에서 스마트폰으로 갈아타기 위해 탈퇴 <br>
     -> ID를 이전해서 피처폰에서 쓰던 계정을 그대로 사용하는 유저<br>
     -> ID를 새로 만들어서 사용하는 유저<br>
     -> 자연히 탈퇴하는 유저<br>
<br>
* 문제 부가 설명<br>
  - 피처폰에서 스마트폰으로 ID 이전을 한 유저는 스마트폰 유저로 간주됨.
  - 스마트폰으로 이전할 때 새로 ID를 만들어서 게임을 이용하는 유저도 있을 것임.
    - 이 경우는 실제 유저가 같더라도 계정이 다르기 때문에 피처폰에서 갈아탄 기존 유저인지 알 방법이 없음.
    - 다만 이 경우는 정기적인 설문조사를 통해 이런 유저수는 극히 일부임이 밝혀짐.
  - 스마트폰으로 갈아탈 때 자연히 게임을 관두는 유저도 있을 것임.
    - 이 경우는 피처폰을 쓰다가 그냥 탈퇴한 것인지, 스마트폰으로 갈아타다보니 탈퇴한 것인지 알 방법이 없음.

  >위 내용을 가지고 '낚시랜드' 게임 기획자 개발자와 회의를 한 결과, '스마트폰으로 갈아타면서 게임을 떠난 유저 중 ID 이전에 실패해서 게임을 떠난 유저가 있지 않을까?' 라는 의견이 있었음.

  >ID 이전 기능은 스마트폰이 처음 나왔을 때 급히 만든 것이라 유저 인터페이스가 조금 어려웠을 것.<br> 그런데도 지금까지 별 문제가 없었던 것은 이 기능이 제공되었을 땐 스마트폰으로 갈아타던 유저들이 소위 '얼리어댑터'라서 유저 인터페이스가 조금 복잡하더라도 알아서 잘 이전했기 때문이라는 추측을 하게됨.

  >만약 피처폰에서 쓰던 계정을 스마트폰으로 이전하려고 했으나 잘 안되어서 게임을 떠난 것이라면, 조속히 시스템을 개선해야됨. 그러나 현재 ID 이전 시스템이 졸속으로 만들어진 관계로 상당한 예산이 필요할 것으로 예상됨. 또한 현재 진행중인 시스템 개선 사안도 있기 때문에 '일단 어떻게든 고치자' 는 식으로 문제를 해결할 수도 없는 상황.

  >일단 대충이라도 좋으니 가급적 빨리 'ID 이전 실패로 인한 탈퇴'가 전체 유저수 감소에 어느 정도 영향을 끼쳤는지 파악해서, 다음과 같은 식으로 액션을 취하기로 함.<br>
  -> 영향이 크다면 다른 사안을 중지해서라도 ID 이전 시스템을 개선한다.<br>
  -> 영향이 작다면 스마트폰 유저 획득을 위한 광고를 개선한다.

- 탈퇴 전의 이용 상황을 생각해보면 다음과 같은 특징이 있을 것.
  - 이용 횟수가 적음
    1. 스마트폰을 가지고 있지 않으며 자연히 게임을 떠남
    2. 스마트폰을 가지고 있으며 자연히 게임을 떠남
  - 이용 횟수가 많음
    3. 스마트폰에서는 새로 ID를 만들어서 게임을 이용함
    4. 스마트폰으로 ID를 이전해서 게임을 이용함
    5. 스마트폰으로 ID를 이전하려 했으나 잘 되지 않았음<br>
>3번은 설문조사를 통해 극히 소수임을 알고 있고, 실제 얻을 수 있는 데이터는 4번 밖에 없음.

## 데이터 수집과 가공
### 분석에 필요한 데이터 수집

```python
#필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
#CSV 파일 읽어 들이기
dau = pd.read_csv('section7-dau.csv')
```

```python
dau.info()        
# region_month : 이용월
# region_day : 이용일
# app_name : 게임명
# user_id : 유저 ID
# device : 단말기 종류
print("============================================")
print(dau)
```

```python
#유저별로 ID 이전을 한 유저인지 아닌지 나타내는 데이터 정리
# FP = 피처폰(Feature Phone) 이용자 , SP = 스마트폰(Smart Phone)이용자
# + 31일치의 데이터 존재로 인한 중복값 제거
#MAU
mau = dau[['user_id','region_month','device']]
mau = mau.drop_duplicates(['region_month', 'device', 'user_id'])
mau=mau.sort_values('user_id')
mau=mau.reset_index(drop=True)

#1월 MAU
mau1=mau[mau['region_month']=='2013-01']
#2월 MAU
mau2=mau[mau['region_month']=='2013-02']

#1월 FP MAU
fp_mau1 = mau1[mau1.device == 'FP']
#2월 FP MAU
fp_mau2 = mau2[mau2.device == 'FP']

#1월 SP MAU
sp_mau1 = mau1[mau1.device == 'SP']
#2월 SP MAU
sp_mau2 = mau2[mau2.device == 'SP']

mau
```

```python
# 1월에 피처폰으로 이용하던 유저가 2월에 탈퇴했는지 혹은 계속 이용했는지 추출
mau2['is_access']=1
fp_mau1=pd.merge(fp_mau1,mau2[['user_id','is_access']],how = 'left', on = ['user_id'])
fp_mau1['is_access']=fp_mau1['is_access'].fillna(0).astype(int)
fp_mau1
```

```python
#1월에 피처폰으로 이용하던 유저가 2월에 이용한 단말기가 피처폰인지 스마트폰인지 구별하기
fp_mau2['is_fp']=1
sp_mau2['is_sp']=1
fp_mau1 = pd.merge(fp_mau1, fp_mau2[['user_id', 'is_fp']],on='user_id', how = 'left')
fp_mau1['is_fp'] = fp_mau1['is_fp'].fillna(0).astype(int)
fp_mau1 = pd.merge(fp_mau1, sp_mau2[['user_id', 'is_sp']],on='user_id', how = 'left')
fp_mau1['is_sp'] = fp_mau1['is_sp'].fillna(0).astype(int)
fp_mau1
```

```python
#로지스틱 회귀분석에 필요한 데이터 추출
#1월에는 피처폰으로 이용했는데 2월에는 이용하지 않았거나 혹은 2월에 스마트폰으로 이용한 유저
fp_mau1 = fp_mau1[(fp_mau1.is_access == 0 ) | (fp_mau1.is_sp == 1)]
fp_mau1
```

```python
#날짜별 게임 이용 상황 데이터 정리하기
fp_dau1 = dau[(dau.device == "FP") & (dau.region_month == "2013-01")]
fp_dau1['is_access'] = 1
fp_dau1 = fp_dau1.pivot_table(index=['user_id'],columns = 'region_day',values = 'is_access').reset_index()
fp_dau1 = fp_dau1.fillna(0).astype(int)
fp_dau1
```

```python
#2월에 스마트폰으로 이용한 유저 데이터 결합하기
#1월에 피처폰으로 이용했던 유저들에 대해, 탈퇴한 유저들은 유저군 '0'으로, 
#스마트폰으로 옮겨서 이용한 유저들은 유저군 '1'로 구분되어 있음
fp_dau_cast = pd.merge(fp_dau1, fp_mau1[['user_id', 'is_sp']], on = ['user_id'], how = 'inner')
fp_dau_cast = fp_dau_cast.fillna(0).astype(int)
fp_dau_cast
```

### 모델 구축과 모델 검증 _ 로지스틱회귀분석

```python
# 로지스틱 회귀 모형 생성
import statsmodels.api as sm
X = fp_dau_cast.drop(['user_id','is_sp'], axis=1)
y = fp_dau_cast['is_sp']

# 상수항 추가
X = sm.add_constant(X)

# 모델 생성
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# 모델 결과 확인
result.summary()
```

```python
# chatGPT로 구현한 로지스틱 회귀 단계적 선택법 (AIC를 기준으로)
def stepwise_selection_aic(X, y, initial_list=[], verbose=True, direction='both'):
    included = list(initial_list)
    best_aic = float('inf')  # 초기 값 설정
    while True:
        changed=False
        excluded = list(set(X.columns)-set(included))
        new_aic = pd.Series(index=excluded)
        for new_column in excluded:
            if direction == 'forward':
                model = sm.Logit(y, sm.add_constant(pd.DataFrame(X[included+[new_column]]))).fit(disp=0)
            else:
                model = sm.Logit(y, sm.add_constant(pd.DataFrame(X[included+[new_column]]))).fit(disp=0)
                if best_aic is None or model.aic < best_aic:
                    direction = 'forward'
                else:
                    direction = 'backward'
                    included.remove(new_column)
            new_aic[new_column] = model.aic
        min_aic = new_aic.min()
        if min_aic < best_aic:
            best_aic = min_aic
            best_feature = new_aic.idxmin()
            included.append(best_feature)
            changed=True
            if verbose:
                print(f'Add {best_feature} with AIC {best_aic:.2f}')
        if not changed:
            break
    while True:
        changed=False
        model = sm.Logit(y, sm.add_constant(pd.DataFrame(X[included]))).fit(disp=0)
        pvalues = model.pvalues.iloc[1:]
        worst_aic = model.aic
        worst_feature = None  # 초기화
        for col in pvalues.index:
            temp = included.copy()
            temp.remove(col)
            model = sm.Logit(y, sm.add_constant(pd.DataFrame(X[temp]))).fit(disp=0)
            if model.aic < worst_aic:
                worst_aic = model.aic
                worst_feature = col
        if worst_feature is not None and worst_aic < model.aic:
            included.remove(worst_feature)
            changed=True
            if verbose:
                print(f'Drop {worst_feature} with AIC {worst_aic:.2f}')
        if not changed:
            break
    return included
```

```python
y_list = stepwise_selection_aic(X, y, direction='both')
print(y_list)
```

```python
y_list.sort()
# 로지스틱 회귀 재실행
X = fp_dau_cast[y_list]
y = fp_dau_cast['is_sp']

# 상수항 추가
X = sm.add_constant(X)
# 모델 생성
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# 모델 결과 확인
print('AIC : ',result.aic)
result.summary()
```

```python
# 잔차 추출
residuals = result.resid_dev
mean = round(np.mean(residuals), 4)
stddev = round(np.std(residuals), 4)
minimum = round(np.min(residuals), 4)
first_quantile = round(np.percentile(residuals, 25), 4)
median = round(np.median(residuals), 4)
third_quantile = round(np.percentile(residuals, 75), 4)
maximum = round(np.max(residuals), 4)
df = pd.DataFrame({'Mean': [mean],
                   'Std. Dev.': [stddev],
                   'Minimum': [minimum],
                   '25th percentile': [first_quantile],
                   'Median': [median],
                   '75th percentile': [third_quantile],
                   'Maximum': [maximum]})
df
```
