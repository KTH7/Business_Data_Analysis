>* 광고의 특징<br>
인터넷 광고는 단가가 쌀 뿐만 아니라 비용에 비해 안정적으로 신규 유저를 획득할 수 있습니다. 비용에 대해 어느 정도까지는 효과를 기대할 수 있지만, TV나 잡지 등의 매스미디어 매체와 달리 광고내용을 전달할 수 있는 고객층 확대에는 한계가 있습니다.<br>
그래서 보통은 인터넷 광고로 출발해서 일정한 유저수를 넘은 시점부터 매스미디어 광고를 시작하는데, 매스미디어 광고는 인터넷 광고에 비해 매우 큰 비용이 듭니다. 또한 광고를 내는 매체에 따라 광고 효과 CPI(Cost Per Install) : 신규 유저 1명을 획득하기 위한 비용이 크게 달라집니다.

## 데이터 수집과 가공
### 분석에 필요한 데이터 수집

```python
#필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
#CSV 파일 읽어 들이기
ad_result = pd.read_csv('ad_result.csv')
```

```python
ad_result.info()
# month : 월
# tvcm : TV 광고비
# magazine : 잡지 광고비
# install :신규 유저수
print("============================================")
print(ad_result)
```

```python
#날짜 형식 변환
import datetime
def date_convert(date_to_convert):
     return datetime.datetime.strptime(date_to_convert, '%b-%y').strftime('%Y-%m')

ad_result['month'] = ad_result['month'].apply(date_convert)
ad_result
```

## 데이터 분석

* 그래프로 데이터를 시각화하기
  -먼저 광고비와 신규 유저수 간에 관계가 있는지 확인할 필요가 있음.<br> 만일 관계가 없다면 광고를 한다고 해서 유저수가 늘어난다고 할 수 없음.

```python
#TV광고의 광고비용과 신규 유저수의 산점도 그리기
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.rcParams['font.family'] ='Malgun Gothic'
# x축, y축에 들어갈 데이터 column 순서대로 지정
plt.subplot(121)
plt.scatter(ad_result['tvcm'], ad_result['install'], c = 'red', alpha = 1)
plt.xlabel('TV 광고비', size = 10)
plt.ylabel('신규 유저수', size = 10)

plt.subplot(122)
plt.scatter(ad_result['magazine'], ad_result['install'], c = 'blue', alpha = 1)
plt.xlabel('잡지 광고비', size = 10)
plt.ylabel('신규 유저수', size = 10)

plt.show()
```

>TV광고와 잡지 광고 모두 오른쪽 위로 올라가는 모습으로 점들이 찍힌 것을 확인할 수 있으며, 이를 통해 광고비를 늘리면 신규 유저수도 늘어나는 관계가 있다고 생각할 수 있음<br>

>이처럼 데이터 간에 상관이 있을 것으로 판단되었으므로 바로 회귀분석을 실시.

### 회귀분석 실행

```python
import statsmodels.api as sm
X = ad_result[['tvcm', 'magazine']]
y = ad_result['install']

# 모델 생성
X = sm.add_constant(X) # 상수항 추가
model = sm.OLS(y, X) # OLS 모델 생성
result = model.fit()

# 모델 결과 확인
result.summary()
```

### 회귀분석의 결과 해석
>신규 유저수 = 0.136 * TV 광고비 + 0.725 * 잡지 광고비 + 188.174<br>

- 신규 유저는 광고를 실시하지 않을 때 월 188명 정도임을 알 수 있음
- TV 광고에 1(1만원)을 투입하면 약 0.1명의 신규 유저를 획득할 수 있음
- 잡지 광고에 1을 투입하면 약 0.7명의 신규 유저를 획득할 수 있음<br>
>즉, 잡지 광고가 TV 광고보다 훨씬 효율적으로 유저를 획득할 수 있다.

#### 더 자세한 정보

```python
# Residuals
# 잔차(예측값과 측정값의 차이)분포를 사분위수로 표현하여 유별나게 어느 한쪽으로 치우쳐 있는지 판단할 수 있음. 
# 이론적으로 잔차의 평균은 항상 0임. 
# 0이 아닌 경우 모델의 설명력이 약한 것으로 판단할 수 있으며, 추가적인 조사가 필요할 수 있음.

# 잔차 추출
residuals = result.resid
mean = round(np.mean(residuals), 4)
stddev = round(np.std(residuals), 4)
minimum = round(np.min(residuals), 4)
first_quantile = round(np.percentile(residuals, 25), 4)
median = round(np.median(residuals), 4)
third_quantile = round(np.percentile(residuals, 75), 4)
maximum = round(np.max(residuals), 4)
df = pd.DataFrame({'Mean': [mean],
                   'Std. Dev.': [stddev],
                   'Minimum': [minimum],
                   '25th percentile': [first_quantile],
                   'Median': [median],
                   '75th percentile': [third_quantile],
                   'Maximum': [maximum]})
df
```

```python
# Coefficients
# 추정된 절편과 기울기에 관한 개요. 
# 각 행은 추정값, 표준오차, t값, p값 순으로 나열되어 있으며 기울기가 얼마나 변동하는지, 통계적 유의성이 있는지 알 수 있음
tables = result.summary().tables[1]
tables
```

```python
#R-squared, Adjusted R-squared
# 결정계수와 자유도 조정 결정계수로 결정계수는 1에 가까워 질수록 모델이 주어진 데이터에 적합함을 의미함
R_Squared = float(result.summary().tables[0].data[0][3])
Adj_R_squared = float(result.summary().tables[0].data[1][3])
df = pd.DataFrame({'R_Squared': [R_Squared],
                   'Adj_R_squared ': [Adj_R_squared ]})
df
```

## 데이터 분석 결론
>잔차분포를 보면, 1Q(제1사분위수)의 절대값이 3Q(제3사분위수)의 절대값보다 커서 분포에 약간의 치우침이 있어 보이지만, 자유도 결정계수가 0.92로 여전히 높기 때문에, 광고 전략을 위한 의사결정에는 문제가 없어 보임

## 액션
> 데이터 분석 결과 TV 광고보다 잡지 광고가 광고효과가 높다는 것을 알 수 있다.<br> 얼핏 보면 TV 광고를 버리고 잡지 광고만 해도 좋을 것 같지만 이 회귀식은 어디까지나 이번에 얻은 데이터만으로 산출한 것임을 잊지 말아야 한다.<br> 데이터의 범위를 넘어선 값이 회귀식에 들어맞지 않는 일은 흔히 있을 수 있기때문. <br>또한 첫 부분에서 밝혔듯이 각 광고매체와 원만한 관계를 유지하기 위해 적어도 3개월에 한 번은 광고 발주를 해야한다.

```python
ad_result['ad_cost']=ad_result[['tvcm', 'magazine']].sum(axis=1)
print('10개월 평균 광고비:',np.mean(ad_result['ad_cost']))
ad_result['ad_CPI']=round((ad_result['ad_cost']/ad_result['install'])*10000,1)
print('10개월 평균 CPI:',np.mean(ad_result['ad_CPI']))
```

>결과적으로 다음과 같이 광고를 배분하기로 함<br>
  -> TV 광고 : 42000(만원), 잡지 광고 : 75000(만원)<br>
  - 이는 10개월 평균 광고비인 129150(만원) 보다 117000(만원)으로 12150(만원)의 비용을 절감할 수 있다.
  
  
>그 결과 다음 식에 따라 대략 60,000명의 신규 유저를 획득할 수 있다는 전망을 내놓을 수 있었음<br>
  ->60275명 = 0.136 * 42000(만원) + 0.725 * 75000(만원) + 188<br>
  <br>
- 이 또한 10개월 평균 CPI인 24295원 보다 훨씬 낮은 수치인 19411원으로 CPI를 낮출 수 있고 앞서 목표했던 22820원 보다도 훨씬 낮은 수치이다.
