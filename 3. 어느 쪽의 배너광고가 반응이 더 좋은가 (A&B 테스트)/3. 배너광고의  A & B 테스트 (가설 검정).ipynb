# 배너광고의 A/B 테스트
* 퍼즐컬렉션에서는 매월 게임에서 이용하는 아이템을 세일하고 있습니다. </br>
이 세일은 매상 비율이 높은 중요한 이벤트인데, 어느 날 회사 경영진으로부터 '매상 자체는 높지만 구매율은 그리 높지 않은 것 아닌가?'라는 지적을 받게 되었습니다.</br>
실제로 다른 앱의 세일과 비교한 결과, 아니나 다를까 퍼즐컬렉션의 구매율이 상대적으로 낮음을 알게 되었습니다. 그리하여 구매율이 낮은 원인을 데이터 분석으로 밝혀내어 개선하라는 지시가 내려왔습니다.

## 현실의 모습 / 이상적인 모습
* 현실의 모습 : 다른 앱의 세일과 비교해서 퍼즐컬렉션의 아이템 세일 구매율이 낮다.
* 이상적인 모습 : 다른 앱과 같은 수준의 구매율을 실현하는 것

## 문제 발견
* 문제
  - 다른 앱의 세일과 비교해서 퍼즐컬렉션의 아이템 세일 구매율이 낮다. (사실)
  <br>
* 가설
  1. 아이템 세일내용에 문제가 있음 (가설)
      - 지금 별로 필요하지 않은 아이템을 내놓고 있다.
      - 세일폭이 별로 크지 않아서 유저에게 크게 어필하지 못하고 있다.
  2. 배너광고의 표시내용에 문제가 있음 (가설) <br>  
  <br>
* 타 부서들과 질의응답 결과
  1. 기획부
  - 아이템 세일에 쓸 만한 것들을 내놓았다고 생각하지만 얼마나 사용하고 있는지 크게 자신은 없음...
  - 세일폭은 50%로 다른 앱과 같고 유저 입장에서도 충분히 싸게 느낄 것임
  2. 마케팅부
  - 아이템 세일의 배너광고는 해당 앱의 디자이너가 만들어 앱에 따라 품질이 제각각임
  - 퍼즐컬렉션의 배너광고는 항상 클릭률이 낮음 (퍼즐 컬렉션 클릭율 6%, 앱 A,B는 12%)
  

>실제로 세일 아이템의 이용 현황을 조사한 결과, 세일 기간에 가장 많이 사용되었던 아이템이었음이 밝혀짐. 즉 가설 1로 구매율이 떨어졌을 가능성은 적음</br>  
-> 배너광고의 표시내용에 문제가 있음'이라는 가설을 분석 테마로 선정

## 데이터 수집과 가공
### 데이터 분석 스토리 정리
* 문제
  - 퍼즐컬렉션의 배너광고 클릭률이 다른 앱보다 낮다. (사실)
  - 퍼즐컬렉션의 배너광고 품질에 문제가 있다. (가설)
* 해결책
  - 클릭률이 높은 배너광고를 조사해서 퍼즐컬렉션의 배너광고를 개선한다.
    - 그러나 지금껏 매월 실시해온 아이템 세일의 배너광고를 한 번도 바꾸지 않아 분석을 하려 해도 데이터가 없음.
    - 그래서 2개의 배너광고를 새로 작성해서 어느 쪽이 더 나은지 데이터를 수집하기로 함.
  <br> 
* 기간에 따른 A/B 전후비교로는 외부요인을 배제할 수 없음
  - ex) 배너B를 광고하고 있을 때 전체 구매율이 좋았다.</br>
  배너B를 광고하고 있을 때 다른 이벤트가 크게 히트를 쳤다.</br>
  배너B를 광고하고 있을 때 TV광고가 실시되거나 혹은 방송에 소개되었다.</br>
  배너A를 광고하고 있을 때 ...<br>
  </br>
* A/B 테스트로 외부요인 제거</br>
  </br>
* A와B의 구분은 임의적이어야 함
  - 그룹 A와 B를 구분할 때는 두 그룹 간에 명확한 차이가 나타나지 않게 나눠야함</br>
      ex) A:남성 / B:여성 , A:10~20대 / B:30~40대
  - 두 개의 그룹에서 조건의 차이가 생기지 않도록 하는 가장 단순한 방법은 유저ID를 어떤 숫자로 나눈 나머지로 분류하는 방법</br>
  이 방법을 이용하면 성별,연령대,게임이용시작일 등의 조건이 균등하게 배분될 것.</br>
  </br>
* A/B 테스트를 동시에 여러 개 실시할 때의 함정
  - A/B테스트를 여러 개 동시에 실시할 때 모든 A/B 테스트에서 그룹을 완전히 똑같은 식으로 구분한 경우</br>
    - 이는 완전히 같은 그룹이기 때문에 두 그룹간에 차이가 나타나더라도 그것이 어떤 테스트에 의한 영향인지 알 수 없게됨.
  - 그룹 A와 그룹 B의 유저 비율이 다를 경우</br>
    - 의도치 않게 다른 테스트를 실시 중인 유저 비율이 다르게 나타난 경우.
    - 다른 테스트를 실시 중인 유저가 포함되지 않도록 그룹을 나누어야함.
    - 테스트의 시기를 달리하든지 사전에 A와 B의 KPI 추이를 비교해서 차이가 없음을 확인한 후에 실시해야함.
    - 소셜게임에서 일반적인 KPI는 PV(Page View) : 액세스 횟수, DAU(Daily Active Users) : 1일 이용 유저수, APRU(Average Revenue Per User) : 유저 1명의 월간 평균 지출 금액, 지속율, 과금율 등이 있음.</br>
    </br>
* 통계적 가설검정은 커트라인
  - 데이터가 많은 경우에는 대부분 통계적으로 유의한 차이가 있다라는 결과가 나옴.
  - 가설검정에서 차이가 났기 때문에 두 그룹에는 차이가 있다라고 결론짓지 않고, 적어도 가설검정에서는 의미가 있는 차이가 나타났으므로 이게 비즈니스 상에서도 의미가 있는 차이인지 검토하자라는 식으로 접근해야 함.

>* 분석에 필요한 데이터 검토
  - ab_test_imp : 배너광고의 표시횟수정보
  - ab_test_goal : 배너광고의 클릭횟수정보

### 분석에 필요한 데이터 수집

```python
#필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
#CSV 파일 읽어 들이기
ab_test_imp = pd.read_csv('section5-ab_test_imp.csv')
ab_test_goal = pd.read_csv('section5-ab_test_goal.csv')
```

```python
ab_test_imp.info()
# log_date : 표시된 날짜
# app_name : 앱 이름
# test_name : 테스트 이름
# test_case : 테스트 케이스: A 또는 B
# user_id : 유저 ID
# transaction_id : 트랜잭션 ID
print("============================================")
print(ab_test_imp)
```

```python
ab_test_goal.info()
# log_date : 클릭한 날짜
# app_name : 앱 이름
# test_name : 테스트 이름
# test_case : 테스트 케이스: A 또는 B
# user_id : 유저 ID
# transaction_id : 트랜잭션 ID
print("============================================")
print(ab_test_goal)
```

### 데이터 가공(전처리)

```python
# abtest_imp에 abtest_goal 결합하기
ab_test = pd.merge(ab_test_imp, ab_test_goal, on=['transaction_id'],how='left', suffixes=("",".g"))
ab_test=ab_test.sort_values(by='transaction_id')
ab_test=ab_test[['transaction_id','log_date','app_name','test_name','test_case','user_id','user_id.g']]
ab_test=ab_test.reset_index(drop=True)
ab_test
```

```python
#클릭 플래그 추가 (클릭 했으면 1 , 안했으면 0)
#'user_id.g컬럼이 'NaN값이면 0으로 그렇지 않으면 1로 재설정
ab_test['is_goal'] = ab_test['user_id.g'].fillna(0).astype(int)
ab_test.loc[ab_test.is_goal > 0, 'is_goal'] = 1
ab_test=ab_test.drop('user_id.g', axis=1)
ab_test
```

```python
#클릭률 집계하기
ab_test_isgoal = ab_test.groupby(['test_case','is_goal'])['is_goal'].count()
df=ab_test_isgoal.unstack()
df['cvr'] = (df[1] / (df[0] + df[1]))
df
```

>A의 클릭률 약 8%</br>
B의 클릭률 약 12% </br>
깊게 들여다 보기전에 가설 검정 실시.

```python
#카이제곱 검정 실행
chisquare = np.array([[3542, 40592], [5056, 38734]])
from scipy.stats import chi2_contingency
chi2, pval, dof, expected = chi2_contingency(chisquare)
# print("chi2:", chi2)
print("p-value:", pval)
# print("dof:", dof)
# print("expected values:\n", expected)
```

>카이제곱 검정을 실시한 결과, p-value값이 0.05보다 작으므로 통계적 차이가 있다고 말할 수 있다.</br>
즉, 배너광고가 A냐 B냐에 따라 클릭률이 변했다고 할 수 있다.<br>

>통계적으로 유의한 차이가 있는지 확인했기에 이것이 비즈니스상에서도 의미있는 차이인지 확인해보자.</br>


>아까 [퍼즐컬렉션의 배너광고는 항상 클릭률이 낮음 (퍼즐 컬렉션 클릭율 6%, 앱 A,B는 12%)]에서 다른 앱은 12%라서 그 차이를 좁히고자 A/B 테스트를 진행했던 것이고, 배너광고 B의 클릭률은 약 12%로 목적을 거의 달성했으므로 이는 비즈니스 상에서 의미있는 차이라고 할 수 있다.

>클릭률의 시계열 추이를 확인해보자.</br>
시계열로 보더라도 전체적으로 B의 클릭률이 높으면 문제가 안되지만, 혹시 어떤 특징 시점에만 차이가 크게 나타난다고 하면 전혀 다른 요인일 가능성이 있다.

```python
#테스트 케이스별로 클릭률 산출
#날짜별, 테스트 케이스별로 클릭률 산출
ab_series = ab_test[['log_date', 'test_case','user_id','is_goal']]
#날짜별, 테스트 케이스 별로 user_id수 카운트해서 넣기
user_count = ab_series[['log_date', 'test_case', 'user_id']]
user_count = user_count.groupby(['log_date', 'test_case']).count()

#날짜별, 테스트 케이스 별로 isgoal수의 합계 넣기
isgoal_count = ab_series[['log_date', 'test_case', 'is_goal']]
isgoal_count = isgoal_count.groupby(['log_date', 'test_case']).sum()

#인덱스 생성
user_count = user_count.reset_index()
isgoal_count = isgoal_count.reset_index()

# #merge로 데이터 병합
ab_final = pd.merge(user_count, isgoal_count,on=['log_date','test_case'])

# #컬럼 이름변경
ab_final.rename(columns = {'user_id':'imp', 'is_goal':'cv'},inplace=True)

# #cvr 컬럼 추가
ab_final['cvr'] = ab_final['cv']/ab_final['imp']

# #최종 데이터
ab_final=ab_final.sort_values(by=['test_case','log_date'])
ab_final
```
