# 소셜게임의 유저 세그먼테이션
>매스미디어 광고에 힘입어 순조롭게 유저수를 늘려가던 퍼즐컬렉션이었는데, 최근 수개월 동안 유저수가 좀처럼 늘지 않고 있습니다.<br> 지금까지는 신규 유저 수가 많았기 때문에 게임운영도 신규 유저를 타깃으로 정한 것이 많았는데, 앞으로는 기존 유저를 위한 대책도 마련하고자 한다며 게임기획자와 경영자로부터 연락이 왔습니다.<br>

>기존 유저를 위한 대책을 실시하기 위해서는 기존 유저가 어떤 유저들인지 이해하는 것이 매우 중요합니다.

* 세그먼테이션<br>
  시장이 어떤 사람들로 구성되어 있는지 이해하는 방법으로 세그먼테이션이 있지만, 기존 세그먼테이션은 단순히 유저의 속성정보를 사용하는 것으로 '우리회사의 주 타깃 고객층은 20대 여성'과 같은 식입니다.<br>
  이런 세그먼테이션으로는 큰 단위로 분류할 수밖에 없기 때문에, 유저의 기호를 파악하기 힘들며 따라서 대책을 세우기도 어렵습니다.<br>
  아울러 퍼즐컬렉션이라는 게임을 즐기고 있다는 사실만으로도 게임기획 당시에 설정했던 타깃 고객층일 확률이 높고, 그로 인해 세그먼테이션을 해도 당연한 결과밖에 얻지 못할 가능성이 있습니다.<br>
  예를 들어 게임기획 시 '20대 남성'을 타깃 유저층으로 잡았다면 위와 같이 세그먼테이션을 해서 '20대 남성이 많음'이라는 결과를 내놓더라도 '당연하지. 그걸 말이라고 하나?' 라는 소리를 들을 뿐입니다.
## 현실의 모습 / 이상적인 모습
* 현실의 모습 : 타깃 유저층이 명확하지 않은 상태
* 이상적인 모습 : 타깃 유저층이 명확한 상태

이번 사례에서 유저 세그먼테이션은 단순히 유저를 분류하는 것으로 끝나는 것이 아니라, 이후에 구체적인 대책을 세울 수 있도록 분류해야 할 필요가 있습니다.<br>
앞서 설명한 '20대 초반의 남성'과 같은 세그먼테이션으로는 어떤 대책을 세워야 할 지 논의할 수 없습니다.<br>
따라서 게임상의 행동패턴으로 세그먼테이션을 실시하는 방법을 생각해봅시다.<br>
예를 들어, 다음과 같습니다.
  - 배틀횟수가 많은 사람
  - 다른 유저에게 배틀을 자주 거는 사람
  - 다른 유저와 잘 협력하는 사람
  - 메시지를 자주 주고받는 사람<br>

즉 '평소에 주로 취하는 행동으로 그 유저의 기호나 성향을 이해할 수 있지 않을까'라고 접근해보는 것입니다.<br>
또한 각 세그먼트의 KPI를 조사해서, KPI가 높은 세그먼트와 낮은 세그먼트의 행동에 어떤 차이가 있는지 알 수 있습니다.<br>
그 결과를 가지고 KPI가 낮은 세그먼트의 유저들에게 어떤 식으로 접근해서 KPI를 늘릴 수 있을지 검토할 수 있습니다.

* 타 부서들과 미팅 결과
  - 기획부
    - 그런 세그먼테이션 방법이라면 각 유저층에 맞는 대책을 세울 수 있을 듯하다.<br>
  - 미팅을 진행하는 동안 다음과 같은 새로운 관점을 발견할 수 있었음.
    1. 게임 내에서 실시되고 있는 이벤트의 참여도에 따라서도 유저층이 크게 다르다.
    2. 랭킹 상위 유저층이 중요하기 때문에 그곳부터 살펴봤으면 한다.

열심히 이벤트에 참여하는 유저와 그렇지 않은 유저는 큰 차이점이 있을 것입니다.<br>
또한 퍼즐컬렉션의 매상 대부분은 랭킹 상위 유저들에 의한 것이기 때문에, 이 유저층을 이해하는 것은 매우 중요합니다.

>따라서 랭킹 상위 유저를 대상으로 행동패턴을 통해 세그먼테이션을 실시하겠습니다.
### 분석 스토리 정리
* 문제
  - 최근 수개월 동안 유저수가 늘지 않고 있음. (사실)
* 사실
  - 랭킹 상위 유저층의 매상이 전체 매상에서 차지하는 비율이 높다. (사실)
* 가설 
  - 행동패턴을 통해 '배틀을 좋아하는 유저', '대화를 좋아하는 유저' 와 같이 취향이나 성향에 따라 세그먼테이션을 할 수 있다. (가설)
* 해결책
  - 각 세그먼테이션에 맞도록 대책을 세워서 실시한다. (해결책)

```python
#필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import warnings
import os
warnings.filterwarnings(action='ignore')
#TSV 파일 읽어 들이기
action=pd.read_csv('./section8_data/daily/action/game-01/2013-10-31/data.tsv',sep='\t')
```

```python
# dau 폴더에 있는 모든 tsv 파일 경로를 리스트로 저장합니다.
tsv_files = []
for root, dirs, files in os.walk('./section8_data/daily/dau'):
    for file in files:
        if file.endswith('.tsv'):
            tsv_files.append(os.path.join(root, file))

# tsv 파일들을 읽어서 데이터프레임으로 만듭니다.
df_list = []
for file in tsv_files:
    df = pd.read_csv(file, sep='\t')
    df_list.append(df)

# 데이터프레임들을 합쳐서 하나의 데이터프레임으로 만듭니다.
dau = pd.concat(df_list, ignore_index=True)
```

```python
# dpu 폴더에 있는 모든 tsv 파일 경로를 리스트로 저장합니다.
tsv_files = []
for root, dirs, files in os.walk('./section8_data/daily/dpu'):
    for file in files:
        if file.endswith('.tsv'):
            tsv_files.append(os.path.join(root, file))

# tsv 파일들을 읽어서 데이터프레임으로 만듭니다.
df_list = []
for file in tsv_files:
    df = pd.read_csv(file, sep='\t')
    df_list.append(df)

# 데이터프레임들을 합쳐서 하나의 데이터프레임으로 만듭니다.
dpu = pd.concat(df_list, ignore_index=True)
```

```python
# dau와 dpu데이터를 log_date와 user_id를 키로 결합
# 비과금 레코드에 0 넣기 , 과금 플래그 추가
dau_dpu=pd.merge(dau,dpu[['log_date','user_id','payment']],on=['log_date','user_id'],how='left')
dau_dpu['payment']=dau_dpu['payment'].fillna(0).astype(int)
dau_dpu['is_payment']=0
dau_dpu.loc[dau_dpu.payment > 0, 'is_payment'] = 1
dau_dpu
```

```python
# 월별, 유저별로 총과금액, 이용일수 집계하기
dau_dpu['log_date']=pd.to_datetime(dau_dpu["log_date"])
dau_dpu['log_date']=dau_dpu['log_date'].dt.strftime('%Y-%m')
# user_id별 월별 payment 합 계산 & 월별 user_id 개수 계산
mau = dau_dpu.groupby(['log_date', 'user_id']).agg({'payment': 'sum','user_id': 'count' }).rename(columns={'user_id': 'access_days'}).reset_index()
mau
```

```python
# 이번 분석은 '랭킹 상위 유저'를 대상으로 하고 있음.
# '랭킹 상위 유저'를 임의로 정하지 않고 데이터로부터 그 기준을 도출하고자 함.
# A47항목이 랭킹 포인트
# 랭킹포인트 분포 확인하기
# 1장에서 소개한 '멱법칙 분포'와 같음
# 랭킹포인트의 분포로부터 유저를 라이트 유저, 미들 유저, 헤비 유저로 분류해서 '헤비 유저'와 '미들 유저'를 '랭킹 상위 유저'로 정의하기로 함.
import matplotlib.pyplot as plt

# 데이터 생성
x = np.arange(len(action['A47']),0,-1)
y = action['A47'].sort_values()

# 꺾은선 그래프 그리기
plt.plot(x, y)

# x, y 축 레이블 지정
plt.xlabel('user')
plt.ylabel('Ranking Point')

# 그래프 제목 지정
plt.title('Ranking Point')

# 그래프 표시하기
plt.show()
```

```python
# 랭킹 포인트로 클러스터링 한 후, 헤비와 미들 클러스터만 뽑아내기
from sklearn.cluster import KMeans
import numpy as np

# 데이터 생성
X = action[['A47','A47']]

# KMeans 모델 생성 및 학습
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# 클러스터 예측
labels = kmeans.predict(X)

# 클러스터링 결과 출력
labels=pd.Series(labels)
print(labels.value_counts())
# 2 : 헤비유저 , 1 : 미들유저, 0 : 라이트 유저
action['labels']=labels
action
```

```python
# 헤비유저와 미들유저만 추출
main=action[action['labels']>=1]
main=main.reset_index(drop=True)

# K-means 알고리즘이 잘 수행되게 대부분의 값이 0인 변수와 상관이 높은 변수 제외하기
sub=main.iloc[:,4:]
sub=sub.iloc[:,:-1]
from sklearn.feature_selection import VarianceThreshold

# nearZeroVar 함수
def near_zero_var(df):
    selector = VarianceThreshold()
    selector.fit(df)
    # 분산이 0인 특성을 제외
    nonzero_ind = np.nonzero(selector.variances_)[0]
    # 결과 반환
    return df.iloc[:, nonzero_ind]

# 데이터 프레임에 적용
user_action_f_filterd = near_zero_var(sub)

# 상관분석 후 highly.cor.f 구하기
user_action_cor = user_action_f_filterd.corr()
highly_cor_f = set()
for i in range(len(user_action_cor.columns)):
    for j in range(i):
        if abs(user_action_cor.iloc[i, j]) > 0.6:
            colname = user_action_cor.columns[i]
            highly_cor_f.add(colname)

# highly.cor.f를 제외한 데이터프레임 만들기
user_action_f_filterd = user_action_f_filterd.loc[
    :, ~user_action_f_filterd.columns.isin(highly_cor_f)]
user_action_f_filterd=sub[["A2","A5","A6","A8","A11","A12","A13","A14","A17","A20","A25","A43","A44","A45","A47","A50","A51","A53","A54"]]
```

```python
# 아직 약한 상관이 남아있음
# 이대로 실행해도 어느 정도 동작은 하지만, 설명변수는 직교인 것이 바람직함
# 이를 위해 주성분 분석을 실행
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
X=user_action_f_filterd.copy()
# 데이터 전처리 - 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 주성분 분석
pca = PCA(n_components=19) # 주성분 추출
pca.fit(X_scaled)

# 주성분 점수 확인
scores = pd.DataFrame(pca.transform(X_scaled), columns=['PC'+str(i+1) for i in range(X_scaled.shape[1])])
scores
```

```python
# 클러스터를 몇개로 나누어야 하는가
# 이번 사례에서는 '어떤 종류의 유저들이 있는가?'를 이해하는 것이 목적이므로, 클러스터 숫자가 너무 많으면 해석하기 힘들어짐
# 몇 개까지라면 해석이 힘들지 않을까? 라는 관점으로 기획담당자에게 확인했고, 3~6개 정도의 범위 내에서 정하기로 함
# 클러스터링
sse = []
for k in range(3,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scores.iloc[:,:1])
    sse.append(kmeans.inertia_)
    
# 엘보우 기법으로 최적 클러스터 수 결정
diff = np.diff(sse)
plt.plot(range(3,11), sse, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()

# 최적 클러스터 수는 5로 정함
```

```python
# k-means 알고리즘 적용
kmeans = KMeans(n_clusters=5,random_state=10).fit(scores.iloc[:,:1])

# 각 데이터가 속한 클러스터 확인
labels = pd.Series(kmeans.labels_)
labels.value_counts()
```

```python
# 클러스터 시각화
clusters = pd.concat([scores, labels.rename("cluster")], axis=1)
# 클러스터별 색깔 지정
colors = ['red', 'blue', 'green', 'purple', 'orange']

# 산점도 그리기
plt.scatter(clusters['PC1'], clusters['PC2'], c=[colors[i] for i in clusters['cluster']])

# 그래프 꾸미기
plt.title('K-means Clustering')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
main['cluster']=clusters['cluster']
radar=main[["A2","A11","A13","A43","A44","A51",'cluster']]
radar = radar.groupby('cluster').mean()
radar=radar.reset_index()
radar
```

```python
# cluster 열을 제외한 열을 표준화
scaler = StandardScaler()
scaled_values = scaler.fit_transform(radar.iloc[:, 1:])

# 표준화된 값으로 데이터프레임 생성
radar_scaled = pd.DataFrame(scaled_values, columns=radar.columns[1:])

def create_radar_chart_data_frame(df):
    df_max = np.max(df, axis=0) + 1
    df_min = np.min(df, axis=0) - 1
    normalized_df = (df - df_min) / (df_max - df_min)
    return pd.concat([pd.DataFrame(df_max).transpose(), pd.DataFrame(df_min).transpose(), normalized_df], ignore_index=True)

radar_scaled=create_radar_chart_data_frame(radar_scaled)

# 결과 출력
radar_scaled = radar_scaled.iloc[2:]
radar_scaled=radar_scaled.reset_index(drop=True)
radar_scaled['cluster']=radar['cluster']
radar_scaled
```

```python
## 하나로 합치기
labels = ['레벨', '지원횟수', '지원받은 횟수', '보스를 쓰러뜨린 횟수', '배틀횟수','게임이용횟수']
num_labels = len(labels)
    
angles = [x/float(num_labels)*(2*np.pi) for x in range(num_labels)] ## 각 등분점
angles += angles[:1] ## 시작점으로 다시 돌아와야하므로 시작점 추가
    
my_palette = plt.cm.get_cmap("rainbow", len(radar_scaled.index))
plt.rcParams['font.family'] ='Malgun Gothic'

fig = plt.figure(figsize=(6,6))
fig.set_facecolor('white')
ax = fig.add_subplot(polar=True)
for i, row in radar_scaled.iterrows():
    color = my_palette(i)
    data = radar_scaled.iloc[i].drop('cluster').tolist()
    data += data[:1]
    
    ax.set_theta_offset(np.pi / 2) ## 시작점
    ax.set_theta_direction(-1) ## 그려지는 방향 시계방향
    
    plt.xticks(angles[:-1], labels, fontsize=13) ## 각도 축 눈금 라벨
    ax.tick_params(axis='x', which='major', pad=15) ## 각 축과 눈금 사이에 여백을 준다.
 
    ax.set_rlabel_position(0) ## 반지름 축 눈금 라벨 각도 설정(degree 단위)
    plt.yticks([0,4], fontsize=10) ## 반지름 축 눈금 설정
    plt.ylim(0,1)
    
    ax.plot(angles, data, color=color, linewidth=2, linestyle='solid', label=row.cluster.astype(int)) ## 레이더 차트 출력
    ax.fill(angles, data, color=color, alpha=0.3) ## 도형 안쪽에 색을 채워준다.
    
plt.legend(loc=(0.9,0.9))
plt.show()
```

#### 각 클러스터의 해석
그룹0 : 자신이 강해지고 싶은 유저 -> 지원횟수와 지원받은 횟수가 적으며 게임이용횟수가 많음<br>
그룹1 : 상호 협력적인 유저 -> 지원횟수와 지원받은 횟수가 모두 많음<br>
그룹2 : (상위 유저층에서는 비교적) 라이트 유저 -> 다른 클러스터와 비교해서 전체적으로 낮음<br>
그룹3 : 협력적인 유저 -> 지원횟수가 조금 많음<br>
그룹4 : 배틀만 즐기는 유저 -> 배틀횟수가 많고 다른 지표는 낮음

>클러스터별로 KPI를 조사해보자.

```python
cluster_KPI=pd.merge(main[['user_id','cluster']],mau,on='user_id')
# cluster 별 payment 평균과 access_days 평균 구하기
result = cluster_KPI.groupby('cluster').agg({'payment': lambda x: round(x.mean()), 'access_days': lambda x: round(x.mean())}).rename(columns={'payment': 'ARPU','access_days':'평균이용일수'})
# 결과 출력
result
```

### 클러스터별 KPI
그룹 1 (상호 협력적인 유저)가 ARPU, 평균이용일수 모두 가장  높은 것을 알 수 있다.<br>
그룹 3 (협력적인 유저)도 그룹 0 (자신이 강해지고 싶은 유저)이나 그룹 4 (배틀만 즐기는 유저) 과 비교하면 높게 나타나는 것으로 보아, '협력 요소'기 굉장히 중요하다는 것을 알 수 있다.<br>
그룹 2 (라이트 유저) 의 ARPU가 높은 것으로 보아, 그룹 2는 라이트 유저라기보다는 '상위 랭킹에 진입하기 위해 애쓰기 시작한 유저'가 아닐까라고 다시 해석할 수 있음

## 액션
랭킹 상위 유저를 클러스터링으로 분류해서 다음과 같은 사실을 알아냈습니다.
1. 랭킹 상위 유저층은 크게 다섯개의 세그먼트로 분류할 수 있다.
2. 그 중 '(상호)협력적인 유저'들의 KPI가 높다.
이 분석 결과를 가지고 기획부와 다시 미팅한 결과, 먼저 '협력 요소'를 강화하는 대책을 검토하기로 하였습니다.

'행동'을 기준으로 한 세그먼테이션은 '행동 변화'를 노린 대책을 세웠을 때 그것이 효과가 있는지 확인할 때도 사용할 수 있다.<br>
이번에 실시한 분석을 매일 자동으로 실행시켜 결과를 메일로 보고하도록 설정하고, 이 피드백을 통해 '대책 검토 -> 실행 -> 개선' 의 업무 절차를 확립했습니다.
